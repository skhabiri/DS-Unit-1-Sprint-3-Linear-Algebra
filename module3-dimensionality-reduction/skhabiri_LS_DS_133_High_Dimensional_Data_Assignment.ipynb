# Vector Transformations

## 4.1 Plug the corresponding unit vectors into each function. Use the output vectors to create a transformation matrix.

\begin{align}
p(\begin{bmatrix}x_1 \\ x_2 \end{bmatrix}) = \begin{bmatrix} x_1 + 3x_2 \\2 x_2 - x_1 \\  \end{bmatrix}
\\
\\
q(\begin{bmatrix}x_1 \\ x_2 \\ x_3\end{bmatrix}) = \begin{bmatrix} 4x_1 + x_2 + 2x_3 \\2 x_2 - x_1 + 3x_3 \\ 5x_1 - 2x_3 + x_2  \end{bmatrix}
\end{align}

```
i_hat = np.array([1,0,0])
j_hat = np.array([0,1,0])
k_hat = np.array([0,0,1])

Ap_i = np.array([[i_hat[0]+3*i_hat[1]],
                 [2*i_hat[1]-i_hat[0]]
                 ])
Ap_j = np.array([[j_hat[0]+3*j_hat[1]],
                 [2*j_hat[1]-j_hat[0]]
                 ])
# Ap_j = np.array([j_hat[0]+3*j_hat[1],2*j_hat[1]-j_hat[0]])

Ap = np.concatenate((Ap_i,Ap_j), axis=1)
print(Ap_i)
print(Ap)
np.matmul(Ap,i_hat[:2])
np.matmul(Ap,j_hat[:2])
```

```
Aq_i = np.array([4*i_hat[0]+i_hat[1]+2*i_hat[2],
                 2*i_hat[1]-i_hat[0]+3*i_hat[2],
                 5*i_hat[0]-2*i_hat[2]+i_hat[1]])
print(Aq_i)
Aq_j = np.array([4*j_hat[0]+j_hat[1]+2*j_hat[2],
                 2*j_hat[1]-j_hat[0]+3*j_hat[2],
                 5*j_hat[0]-2*j_hat[2]+j_hat[1]])
Aq_k = np.array([4*k_hat[0]+k_hat[1]+2*k_hat[2],
                 2*k_hat[1]-k_hat[0]+3*k_hat[2],
                 5*k_hat[0]-2*k_hat[2]+k_hat[1]])
Aq_i_col = np.array([Aq_i]).T
Aq_j_col = np.array([Aq_j]).T
Aq_k_col = np.array([Aq_k]).T

Aq = np.concatenate((Aq_i_col,Aq_j_col,Aq_k_col), axis=1)
print(Aq)
```

```
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)

print(national.shape)
national.head()
```
